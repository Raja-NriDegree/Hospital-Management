```python

questions = [
    {
        "question": "who is the father of the python?",
        "options": ["A.  Guido van Rossum", "B. Alexander", "C. Roman ", "D. Edison"],
        "answer": "a"
    },
    {
        "question": "which of these is not a data type?",
        "options": ["A. String", "B. Float", "C. Integer", "D. Python"],
        "answer": "D"
    },
    {
        "question": "Where can we use the Python?",
        "options": ["A. Web Development", "B. Data Science and Data Analysis", "C. Gaming", "D. All of These"],
        "answer": "D"
    },
    {
        "question": "What is the correct Extension of Python files ?",
        "options": ["A. .py", "B. .phhy", "C. .java", "D. .html"],
        "answer": "A"
    },
    {
        "question": "which of these  Python function is used to get User Input in Python?",
        "options": ["A. enter()", "B. input()", "C. get()", "D. read()"],
        "answer": "B"
    },
    {
        "question": "What is the following is not a Looping Statement?",
        "options": ["A. do-while", "B. while", "C. for", "D. if"],
        "answer": "D"
    },
    {
        "question": "Which of these following is not a error in Python?",
        "options": ["A. Name Error", "B. Syntax Error", "C. Type error", "D. None of These"],
        "answer": "D"
    },
    {
        "question": "The giving four Spaces is called _________ Space'?",
        "options": ["A. 1 Tab Space", "B. 1 MB Space", "C. 1 Byte Space", "D. 1 space"],
        "answer": "A"
    },
    {
        "question": "Which of these is following a Logic Gate?",
        "options": ["A. And", "B. Switch", "C. Break", "D. Continue"],
        "answer": "A"
    },
    {
        "question": "In Dictinories what data type is mostly used-?",
        "options": ["A. Strings", "B. Char", "C. Integer", "D. Classes"],
        "answer": "A"
    }
]

def quiz_game():
    print("Welcome to the Quiz Game!")
    print("Answer the questions by typing the letter of the correct option (A, B, C, or D).\n")
    
    score = 0
    question_number = 1
    
    for q in questions:
        print(f"Question {question_number}: {q['question']}")
        
        for option in q["options"]:
            print(option)
        
        answer = input("Your answer: ")
        
        if answer == q["answer"]:
            print("Correct!\n")
            score += 10
        else:
            print(f"Wrong! The correct answer was {q['answer']}.\n")
        
        question_number += 1
    
    print(f"Quiz Over! Your total score is: {score}/100")
    
    if score == 100:
        print("Excellent! Your are awesome!")
    elif score >= 50:
        print("Good!")
    else:
        print("Better luck next time!")

if __name__ == "__main__":
    quiz_game()

```
